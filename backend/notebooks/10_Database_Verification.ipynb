Este notebook permite verificar que la base de datos se ha creado correctamente y consultar las predicciones almacenadas.


## 1. Importar M√≥dulos


```python
import sys
from pathlib import Path
import pandas as pd

# A√±adir src al path
project_root = Path('../').resolve()
sys.path.append(str(project_root / 'src'))

from utils.database import get_db_manager
```

## 2. Inicializar Base de Datos


```python
# Inicializar gestor de BD
db_manager = get_db_manager()

print(f"üìÅ Ubicaci√≥n de la BD: {db_manager.db_path}")
print(f"‚úÖ Base de datos lista: {db_manager.db_path.exists()}")
```

## 3. Verificar Estad√≠sticas


```python
# Obtener estad√≠sticas
stats = db_manager.get_statistics()

print("üìä Estad√≠sticas de la Base de Datos:")
print(f"   Total de predicciones: {stats['total_predictions']}")
print(f"   T√≥xicas: {stats['toxic_count']} ({stats['toxic_percentage']:.2f}%)")
print(f"   No t√≥xicas: {stats['not_toxic_count']} ({stats['not_toxic_percentage']:.2f}%)")
print(f"   Confianza promedio: {stats['average_confidence']:.4f}")
```

## 4. Consultar Predicciones Recientes


```python
# Obtener √∫ltimas 10 predicciones
predictions = db_manager.get_predictions(limit=10)

if predictions:
    print(f"üìù √öltimas {len(predictions)} predicciones:")
    print("\n" + "="*80)
    
    for i, pred in enumerate(predictions, 1):
        print(f"\n{i}. ID: {pred['id']}")
        print(f"   Texto: {pred['text'][:100]}..." if len(pred['text']) > 100 else f"   Texto: {pred['text']}")
        print(f"   Etiqueta: {pred['toxicity_label']}")
        print(f"   Probabilidad t√≥xico: {pred['probability_toxic']:.4f}")
        print(f"   Confianza: {pred['confidence']:.4f}")
        print(f"   Origen: {pred['source']}")
        print(f"   Fecha: {pred['created_at']}")
else:
    print("‚ö†Ô∏è  No hay predicciones en la base de datos.")
    print("\nüí° Para crear predicciones:")
    print("   1. Ejecuta la API: cd api && bash run.sh")
    print("   2. Haz una predicci√≥n: POST /predict")
    print("   3. O analiza un video de YouTube: POST /analyze/youtube")
```
